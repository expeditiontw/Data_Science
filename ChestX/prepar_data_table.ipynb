```python

label_name = ['Atelectasis' , 'Cardiomegaly', 'Effusion'
              , 'Infiltration', 'Mass', 'Nodule'
              , 'Pneumonia', 'Pneumothorax', 'Consolidation'
              , 'Edema', 'Emphysema', 'Fibrosis'
              , 'Pleural_Thickening', 'Hernia', 'None']

dir_scr_train = '/data/private/noah/medi_image/data/label/train/'
dir_scr_test = '/data/private/noah/medi_image/data/label/valid/'
dir_name = '/data/private/noah/medi_image/data/binary'

for label in label_name[:-1]:
    tmp_dir = os.path.join(dir_name,label)
    
    tmp_tr_dir = os.path.join(tmp_dir, 'train')
    tmp_ts_dir = os.path.join(tmp_dir, 'test')
    try :
#        os.mkdir(tmp_tr_dir)
#        os.mkdir(tmp_ts_dir)
        os.mkdir(os.path.join(tmp_tr_dir,label))
        os.mkdir(os.path.join(tmp_tr_dir,'None'))
        os.mkdir(os.path.join(tmp_ts_dir,label))
        os.mkdir(os.path.join(tmp_ts_dir,'None'))
        print ('Dir ',label,'created')
        
        src_dir = os.path.join(dir_scr_train,label)
        onlyfiles = [f for f in listdir(src_dir) if isfile(join(src_dir, f))]

        for fname in onlyfiles:
            scr_fname = os.path.join(src_dir, fname)
            dst_fname = os.path.join(tmp_tr_dir,label, fname)
            shutil.copyfile(scr_fname, dst_fname)
            
        src_dir = os.path.join(dir_scr_train,'None')
        onlyfiles = [f for f in listdir(src_dir) if isfile(join(src_dir, f))]

        for fname in onlyfiles:
            scr_fname = os.path.join(src_dir, fname)
            dst_fname = os.path.join(tmp_tr_dir,'None', fname)
            shutil.copyfile(scr_fname, dst_fname)

            
        src_dir = os.path.join(dir_scr_test,label)
        onlyfiles = [f for f in listdir(src_dir) if isfile(join(src_dir, f))]

        for fname in onlyfiles:
            scr_fname = os.path.join(src_dir, fname)
            dst_fname = os.path.join(tmp_ts_dir,label, fname)
            shutil.copyfile(scr_fname, dst_fname)
            
        src_dir = os.path.join(dir_scr_test,'None')
        onlyfiles = [f for f in listdir(src_dir) if isfile(join(src_dir, f))]

        for fname in onlyfiles:
            scr_fname = os.path.join(src_dir, fname)
            dst_fname = os.path.join(tmp_ts_dir,'None', fname)
            shutil.copyfile(scr_fname, dst_fname)
        
    except :
        print ('Dir existed!!')
        
```
